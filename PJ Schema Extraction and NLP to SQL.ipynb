```python
import pandas as pd
import numpy as np
```

```python
fileLevel = pd.read_csv("C:\\ML\\Ramu\\FileLevel_Schema.csv" ,header=None, skiprows=1,names=['combinedata']);

print(fileLevel)
```

```python
# Function to clean and split data
def clean_and_split(row_data):
    if pd.isna(row_data) or not isinstance(row_data, str):
        print(row_data)
        return None, None, None, None
    # Remove surrounding parentheses
    row_data = row_data.strip('()')
    #print(row_data)
    
        # Split data at the first comma
    first_split = row_data.split(',', 1)  # Split only on the first comma
    identifier = first_split[0].strip()

    # The rest (without surrounding quotes) needs further processing
    remaining_data = first_split[1] if len(first_split) > 1 else ''
    
    # Split again to separate schema and table from descriptions
    parts = remaining_data.rsplit(',', 2)  # Split based on the last two commas for schema and table

    # Ensure validity
    if len(parts) < 3:
        return identifier, None, None, None
    
    # Extract middle descriptions and last two schema and table
    description = parts[0].strip().strip('"')  # Middle description part
    schema_name = parts[1].strip()  # Schema name
    table_name = parts[2].strip()  # Table name

    return identifier, description, schema_name, table_name
```

```python
# Applying the function to the DataFrame
fileLevel[['Identifier', 'Description', 'SchemaName', 'TableName']] = fileLevel['combinedata'].apply(lambda x: pd.Series(clean_and_split(x)))
```

```python
columnDesc = pd.read_csv("C:\\ML\\Ramu\\ColumnDesc_Schema.csv" ,header=None, skiprows=1,names=['ColumnDesc']);

print(columnDesc)
```

```python
# Function to clean and split data
def clean_and_split_col_desc(row_data_column):
    if pd.isna(row_data_column) or not isinstance(row_data_column, str):
        print(row_data_column)
        return None, None, None, None
    # Remove surrounding parentheses
    row_data_column = row_data_column.strip('()')
    #print(row_data)
    
        # Split data at the first comma
    first_split = row_data_column.split(',', 1)  # Split only on the first comma
    identifier = first_split[0].strip()
    
    # The rest (without surrounding quotes) needs further processing
    column_desc = first_split[1]
    
    return identifier, column_desc
```

```python
# Applying the function to the DataFrame
columnDesc[['column_name', 'column_desc']] = columnDesc['ColumnDesc'].apply(lambda x: pd.Series(clean_and_split_col_desc(x)))
```

```python
# Drop the original column
columnDesc.drop(columns=['ColumnDesc'], inplace=True)
columnDesc
```

```python
columnlevel = pd.read_csv("C:\\ML\\Ramu\\ColumnLevel_Schema.csv" ,header=None, skiprows=1,names=['columnlevel']);

print(columnlevel)
```

```python
# Function to clean and split data
def clean_and_split_col_level(row_data_column):
    if pd.isna(row_data_column) or not isinstance(row_data_column, str):
        print(row_data_column)
        return None, None, None, None
    # Remove surrounding parentheses
    row_data_column = row_data_column.strip('()')
    #print(row_data)
    
        # Split data at the first comma
    first_split = row_data_column.split(',', 3)  # Split only on the first comma
    identifier = first_split[0].strip()
    
    # The rest (without surrounding quotes) needs further processing
    colName = first_split[1]
    dataType = first_split[2]
    
    return identifier, colName, dataType
```

```python
# Applying the function to the DataFrame
columnlevel[['Identifier', 'column_name','dataType']] = columnlevel['columnlevel'].apply(lambda x: pd.Series(clean_and_split_col_level(x)))
```

```python
# Drop the original column
columnlevel.drop(columns=['columnlevel'], inplace=True)
columnlevel
```

```python
# Merging df1 and df3 on Identifier
merged_df_1_3 = pd.merge(fileLevel, columnlevel, on='Identifier')

# Merging the result with df2 on column_name
final_merged_df = pd.merge(merged_df_1_3, columnDesc, on='column_name')

# Display the final merged DataFrame
print(final_merged_df)
```

```python
# Group by 'SchemaName', 'TableName', and 'Description' to collect column names and descriptions for each table
grouped = final_merged_df.groupby(['SchemaName', 'TableName', 'Description'])

# Create a list of table schema strings
schema = []

for (schema_name, table_name, table_desc), group in grouped:
    # Extract column names and descriptions, formatting each as "name (description)"
    columns_with_desc = ', '.join(
        f"{row['column_name']} ({row['column_desc']})"
        for _, row in group.iterrows()
    )
    
    # Include SchemaName.TableName and format the final string for each table
    table_schema = f"Table: {schema_name}.{table_name} ({table_desc}) ({columns_with_desc})"
    
    # Append to the schema list
    schema.append(table_schema)

# Printing the result
print(schema)
```

```python
from sentence_transformers import SentenceTransformer
import faiss
import numpy as np

# Load embedding model
model = SentenceTransformer("all-MiniLM-L6-v2")  # Small & efficient

# Sample Database Schema
schema = schema

# Convert schema descriptions to embeddings
schema_vectors = model.encode(schema)

# Store embeddings using FAISS
dimension = schema_vectors.shape[1]
index = faiss.IndexFlatL2(dimension)  # L2 distance (cosine similarity)
index.add(np.array(schema_vectors))

# Save schema texts for reference
schema_map = {i: schema[i] for i in range(len(schema))}
```

```python
def find_relevant_schema(query, top_k=2):
    query_vector = model.encode([query])
    
    # Search FAISS index
    _, indices = index.search(np.array(query_vector), top_k)
    # Retrieve relevant schema items
    relevant_schemas = [schema_map[i] for i in indices[0]]
    return relevant_schemas

# Example query
query = "Show me all transactions greater than $1,000 from last week."
relevant_schema = find_relevant_schema(query)
print("Relevant Schema:", relevant_schema)
```
