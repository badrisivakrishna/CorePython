````python
# Function to clean and split data
def clean_and_split(row_data):
    if pd.isna(row_data) or not isinstance(row_data, str):
        print(row_data)
        return None, None, None, None
    # Remove surrounding parentheses
    row_data = row_data.strip('()')
    #print(row_data)
    
        # Split data at the first comma
    first_split = row_data.split(',', 1)  # Split only on the first comma
    identifier = first_split[0].strip()

    # The rest (without surrounding quotes) needs further processing
    remaining_data = first_split[1] if len(first_split) > 1 else ''
    
    # Split again to separate schema and table from descriptions
    parts = remaining_data.rsplit(',', 2)  # Split based on the last two commas for schema and table

    # Ensure validity
    if len(parts) < 3:
        return identifier, None, None, None
    
    # Extract middle descriptions and last two schema and table
    description = parts[0].strip().strip('"')  # Middle description part
    schema_name = parts[1].strip()  # Schema name
    table_name = parts[2].strip()  # Table name

    return identifier, description, schema_name, table_name
````

````python
# Applying the function to the DataFrame
fileLevel[['Identifier', 'Table_Description', 'SchemaName', 'TableName']] = fileLevel['combinedata'].apply(lambda x: pd.Series(clean_and_split(x)))
````

````python
def generate_sql_ollama(question, collection, model="mistral", top_n=3):
    question_embedding = get_embedding_ollama(question)
    if not question_embedding:
        return "Could not generate embedding for the question."

    results = collection.query(
        query_embeddings=[question_embedding],
        n_results=top_n,
        include=['metadatas']
    )

    relevant_schema_info = results['metadatas'][0] # Assuming top result is most relevant
    # You might want to process multiple results for more context

    prompt = f"""
    Given the following database schema information:
    {relevant_schema_info}

    Generate a SQL query to answer the question: "{question}"
    """

    url = "http://localhost:11434/api/chat"
    data = {
        "model": model,
        "messages": [
            {"role": "system", "content": "You are a helpful SQL query generator based on the provided schema."},
            {"role": "user", "content": prompt}
        ],
        "stream": True  # Enable streaming
    }
    try:
        with requests.post(url, headers={'Content-Type': 'application/json'}, json=data, stream=True) as response:
            response.raise_for_status()
            full_response = ""
            for line in response.iter_lines(decode_unicode=True):
                if line:
                    try:
                        chunk = json.loads(line)
                        if "message" in chunk and "content" in chunk["message"]:
                            full_response += chunk["message"]["content"]
                        if "done" in chunk and chunk["done"]:
                            break
                    except json.JSONDecodeError as e:
                        print(f"Error decoding JSON chunk: {e}, line: {line}")
                        continue
            return full_response

    except requests.exceptions.RequestException as e:
        return f"Request error: {e}"
    except json.JSONDecodeError as e:
        return f"Initial JSON decode error: {e}"
````

````python
import re

def extract_sql(ollama_output):
    """
    Extracts the SQL query from the Ollama output string.

    Args:
        ollama_output: The string output from Ollama containing the SQL query.

    Returns:
        The extracted SQL query string, or None if no SQL query is found.
    """
    # Use a regular expression to find the SQL query within the code block.
    match = re.search(r"```sql\n(.*?)\n```", ollama_output, re.DOTALL)
    if match:
        return match.group(1).strip()  # Extract and remove leading/trailing spaces
    # Alternative regex (if the ```sql block is not consistently present)
    match = re.search(r"(SELECT.*?)(?:;|$)", ollama_output, re.DOTALL | re.IGNORECASE)
    if match:
        return match.group(1).strip()

    return None  # Return None if no SQL query is found
````

````python
from sqlglot import transpile

def convert_to_postgres(sql_query):
  """
  Converts a SQL query to PostgreSQL dialect using sqlglot.

  Args:
    sql_query: The original SQL query string.

  Returns:
    A string containing the PostgreSQL equivalent of the query,
    or the original query if conversion fails or isn't significantly different.
  """
  try:
    postgres_query = transpile(sql_query, read="auto", write="postgres")
    if postgres_query:
      return postgres_query[0]  # transpile returns a list
    else:
      return sql_query  # Return original if no significant change
  except Exception as e:
    print(f"An error occurred during conversion: {e}")
    return sql_query
````

from sentence_transformers import SentenceTransformer
import faiss
import numpy as np

# Load embedding model
model = SentenceTransformer("all-MiniLM-L6-v2")  # Small & efficient

# Sample Database Schema
schema = schema

# Convert schema descriptions to embeddings
schema_vectors = model.encode(schema)

# Store embeddings using FAISS
dimension = schema_vectors.shape[1]
index = faiss.IndexFlatL2(dimension)  # L2 distance (cosine similarity)
index.add(np.array(schema_vectors))

# Save schema texts for reference
schema_map = {i: schema[i] for i in range(len(schema))}

def find_relevant_schema(query, top_k=2):
    query_vector = model.encode([query])
    
    # Search FAISS index
    _, indices = index.search(np.array(query_vector), top_k)
    # Retrieve relevant schema items
    relevant_schemas = [schema_map[i] for i in indices[0]]
    return relevant_schemas
